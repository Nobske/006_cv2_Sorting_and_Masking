# CODE TO PERFORM COLOR SELECTION

# Import libraries

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

# Read the image / show the shape / display the image

```python
image_read = mpimg.imread("image_lane_c.jpg")
```

```python
image_read.shape
```

```python
plt.imshow(image_read)
```

# Convert the image to gray scale

```python
image_gray = cv2.cvtColor(image_read, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(image_gray)
```

```python
plt.imshow(image_gray, cmap ="gray")
```

```python
image_gray.shape
```

# Make a copy of the image

```python
image_copy = np.copy(image_gray) # image as array copy from np 
```

```python
image_copy.shape
```

# Sort out the white color of the image

```python
image_copy [(image_copy[:] < 250)] = 0 # take any value of the pixel which is below 250 (max is 255) and make it 0 -> black
```

```python
plt.imshow(image_copy, cmap = "gray")
plt.show()
```
